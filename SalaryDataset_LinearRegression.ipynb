```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Salary Data.csv')
df
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
#calculated missing values
df.isna().sum()
```

```python
#analyzing datatype
print(df.dtypes)
```

```python
#seperate x and y
x=df.iloc[:,:-1].values
y=df.iloc[:,-1]
```

```python
#Draw  Scatterplot
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('Year of Exp')
plt.ylabel('Salary')
plt.title('Year of Exp Vs Salary')
```

```python
#converting data in to training data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
#for Linear Regression model there is no need to do normalization
```

```python
#model Creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[3.1]]))
```

```python
#actual_value
y_test
```

```python
##making 3 columns with (predictedValue)y_pred,  y_test(actual_value,    Difference=actualvalue-predicted value)
df1=pd.DataFrame({'Actual_Value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
#scatter plot--
plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_pred,color='blue')
```

```python
#Finding slope of the graph,y=mx+c
print("slope is",model.coef_)
```

```python
print("constant is",model.intercept_)
```

```python
#root mean squared error
from sklearn.metrics import mean_squared_error
import numpy as np
variable=mean_squared_error(y_test,y_pred)
root_error=np.sqrt(variable)
print(root_error)
```

```python
# mean absolute percenage error
from sklearn.metrics import mean_absolute_percentage_error
print("error percentage is",mean_absolute_percentage_error(y_test,y_pred))
#so the model is having above 90 percentage accuracy
```
